A continuació farem un test d'hipòtesis per a cada columna per comprovar la seva distribució.

H0-> La distribució és normal

H1-> La distribució no es normal

```python
from scipy.stats import shapiro

l={}
t = False
for i in df.columns:
  stat, p = shapiro(df[i])
  #print('stat=%.3f, p-value= %.5f' % (stat, p))
  if p > 0.05:
    l[i] = p
    t = True

if t == True:
  print(f'La/les variables que són probable que tinguin una distribució gausiana són les següents:\n\n{l}')
else:
  print('Cap variable té una distribució normal')
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Definim les variables d'entrada i de sortida
X = df.drop('Class', axis=1)
y = df.Class
```

```python
#Dividim les dades en train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

Entrenem les dades separades amb quatre models de classificació per tal de comparar-los més tard.

## Decission Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier().fit(X_train, y_train)
```

## K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
```

## Support Vector Machine

```python
from sklearn import svm
classifier_svm = svm.SVC(kernel ='linear').fit(X_train, y_train)
```

## XG Boost

```python
import xgboost as xgb
classifier_xgb = xgb.XGBClassifier().fit(X_train, y_train)
```

# Exercici 2
Compara els models de classificació utilitzant la precisió (accuracy), una matriu de confiança i d’altres mètriques més avançades.

### Accuracy: percentatje de les prediccions correctes

image.png

```python
y_pred_dtc = dtc.predict(X_test)
dtc_accuracy = metrics.accuracy_score(y_test, y_pred_dtc)
print('Accuracy:', dtc_accuracy)

y_pred_knn = knn.predict(X_test)
knn_accuracy = metrics.accuracy_score(y_test, y_pred_knn)
print('Accuracy:', knn_accuracy)

y_pred_svm = classifier_svm.predict(X_test)
svm_accuracy = metrics.accuracy_score(y_test, y_pred_svm)
print('Accuracy:', svm_accuracy)

y_pred_xgb = classifier_xgb.predict(X_test)
xgb_accuracy = metrics.accuracy_score(y_test, y_pred_xgb)
print('Accuracy:', xgb_accuracy)
```

```python
models = np.array(['Decission Tree', 'KNearest Neighbors', 'Support Vector Machine', 'XGBoost'])
accuracy = np.array([dtc_accuracy, knn_accuracy, svm_accuracy, xgb_accuracy])
```

```python
fig = plt.figure(figsize=(15,5))

plt.title('Accuracy')
sns.barplot(x= models, y=accuracy)
```

## Matriu de confusió

És una matriu en la que en el eix vertical indica la la classificació real dels elements, i en el eix horitzontal, la clasificació predita pel model, de manera que si te un accuracy del 100%, només hi haurà valors en la diagonal de la matriu i en la resta de caselles, zeros.

```python
from sklearn.metrics import  confusion_matrix
dtc_cm = confusion_matrix(y_test, y_pred_dtc)
knn_cm = confusion_matrix(y_test, y_pred_knn)
svm_cm = confusion_matrix(y_test, y_pred_svm)
xgb_cm = confusion_matrix(y_test, y_pred_xgb)

fig = plt.figure(figsize=(10,7))


ax1 = fig.add_subplot(2, 2, 1)
plt.title('dtc')
sns.heatmap(dtc_cm, annot=True)


ax2 = fig.add_subplot(2, 2, 2)
plt.title('knn')
sns.heatmap(knn_cm, annot=True)

ax3 = fig.add_subplot(2, 2, 3)
plt.title('svm')
sns.heatmap(svm_cm, annot=True)

ax4 = fig.add_subplot(2, 2, 4)
plt.title('xgb')
sns.heatmap(xgb_cm, annot=True)

plt.subplots_adjust(hspace=0.3)
```

# Exercici 3
Entrena’ls usant els diferents paràmetres que admeten per tal de millorar-ne la predicció.



```python
from sklearn.model_selection import RandomizedSearchCV

grid =  [{'criterion': ['gini', 'entropy', 'log_ros'],
         'max_features': ['sqrt', 'log2', 'auto'],
         'max_depth': range(2,1000),
         'min_samples_split': range(2,1000),
         'splitter': ['best', 'random']}]

dtc = DecisionTreeClassifier().fit(X_train, y_train)
dtc_bp = RandomizedSearchCV(dtc, grid, cv=5).fit(X_train, y_train)
y_pred_dtc = dtc_bp.predict(X_test)
dtc_accuracy = metrics.accuracy_score(y_test, y_pred_dtc)
print('Accuracy:', dtc_accuracy)
```

```python
grid =  [{'n_neighbors': range(0,100),
         'weights': ['uniforms', 'distance'],
         'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
         'leaf_size': range(0,100),
         'p': range(0,10)}]

knn = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
knn_bp = RandomizedSearchCV(knn, grid, cv=5).fit(X_train, y_train)
y_pred_knn = knn_bp.predict(X_test)
knn_accuracy = metrics.accuracy_score(y_test, y_pred_knn)
print('Accuracy:', knn_accuracy)
```

```python
grid =  [{'C': range(0,10),
         'degree': range(0,10),
         'gamma': ['scale', 'auto'],
         'coef0': range(0,10)}]

classifier_svm = svm.SVC(kernel ='linear').fit(X_train, y_train)     
svm_bp = RandomizedSearchCV(classifier_svm, grid, cv=5).fit(X_train, y_train)
y_pred_svm = svm_bp.predict(X_test)
svm_accuracy = metrics.accuracy_score(y_test, y_pred_svm)
print('Accuracy:', svm_accuracy)
```

```python
grid =  [{'booster': ['gbtree', 'gblinear', 'dart'],
         'verbosity': range(1,10),
         'validate_parameters': ['False', 'True'],
         'nthread': range(1,10),
         'num_feature': range(0,10)}]

classifier_xgb = xgb.XGBClassifier().fit(X_train, y_train)
xgb_bp = RandomizedSearchCV(classifier_xgb, grid, cv=5).fit(X_train, y_train)
y_pred_xgb = xgb_bp.predict(X_test)
xgb_accuracy = metrics.accuracy_score(y_test, y_pred_xgb)
print('Accuracy:', xgb_accuracy)
```

Després d'haber obtingut els millors paràmetres, l'únic model que ha millorat ha estat Knn, la resta tenen la mateixa puntuació.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler

#Estandarització
df[['Alcalinity of ash']] = StandardScaler().fit(df[['Alcalinity of ash']]).transform(df[['Alcalinity of ash']])

#Normalització
columns = ['Total phenols', 'Flavanoids', 'Nonflavanoids phenols', 'OD280/OD315 of diluted wines', 'Proline']
df[columns] = MinMaxScaler().fit(df[columns]).transform(df[columns])

#Robust Scaler
columns = ['Alcohol', 'Malic acid', 'Ash', 'Magnesium', 'Proanthocyanins', 'Color intensity', 'Hue']
df[columns] = MinMaxScaler().fit(df[columns]).transform(df[columns])
```

```python
df
```

```python
X = df.drop('Class', axis=1)
y = df.Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
dtc = DecisionTreeClassifier().fit(X_train, y_train)
knn = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
classifier_svm = svm.SVC(kernel ='linear').fit(X_train, y_train)
classifier_xgb = xgb.XGBClassifier().fit(X_train, y_train)
```

```python
y_pred_dtc = dtc.predict(X_test)
dtc_accuracy = metrics.accuracy_score(y_test, y_pred_dtc)
print('Accuracy:', dtc_accuracy)

y_pred_knn = knn.predict(X_test)
knn_accuracy = metrics.accuracy_score(y_test, y_pred_knn)
print('Accuracy:', knn_accuracy)

y_pred_svm = classifier_svm.predict(X_test)
svm_accuracy = metrics.accuracy_score(y_test, y_pred_svm)
print('Accuracy:', svm_accuracy)

y_pred_xgb = classifier_xgb.predict(X_test)
xgb_accuracy = metrics.accuracy_score(y_test, y_pred_xgb)
print('Accuracy:', xgb_accuracy)
```
